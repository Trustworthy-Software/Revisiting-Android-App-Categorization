#### Imports

```python
# IMPORTS
from   sklearn.metrics  import adjusted_rand_score# IMPORT
from   sklearn.cluster  import KMeans
from   tqdm             import tqdm
import pandas           as pd
import numpy            as np
import ast
import os
```

```python
print("‚ö° START ‚ö°")
```

#### Parameters

```python
# Ground-Truth Dataset
INPUT_PATH  = "../TMP/1c_EbrahimiFeatures.csv"

# Output Path
OUTPUT_PATH = "../1c_EbrahimiClusteringLabels.csv"

TMP_PATH = "../TMP"
if not os.path.exists(TMP_PATH):
    os.makedirs(TMP_PATH)
    print("üìÅüÜï Folder created       :", TMP_PATH)
else:
    print("üìÅ‚úÖ Folder already exists:", TMP_PATH)
```

```python
RANDOM_SEED = 151836
```

### 1. Load Data

```python
appsDF = pd.read_csv(INPUT_PATH, index_col=False)
print("#Ô∏è‚É£ Apps: {}".format(appsDF.shape[0]))

appsDF.head(3)
```

### 3. Clustering - KMeans

```python
def getARIscore(clusteringLabels):
    ariScore = adjusted_rand_score(appsDF["classID"].values, clusteringLabels)
    print("üéØ ARI: {:.4f}".format(ariScore))
    return
```

```python
# # Parameters
# NUM_CLUSTERS = 50
# PARAMETERS (TEST)
NUM_CLUSTERS = 5
```

```python
# Convert the string column to NumPy arrays
appsDF['gloveFeatures'] = appsDF['gloveFeatures'].apply(eval)

# Get the features
featuresArray = np.array(appsDF['gloveFeatures'].to_list())
```

#### Training

```python
# Apply KMeans clustering to probs_matrix, with the same seed
model = KMeans(n_clusters = NUM_CLUSTERS, random_state = RANDOM_SEED, n_init='auto').fit(featuresArray)

# Get the labels
clusteringLabels = model.predict(featuresArray)

print("\n‚≠ê Ebrahimi et al.")
getARIscore(clusteringLabels)
```

### 3. Save everything

```python
# Save the labels into the Pandas DF
appsDF = appsDF.loc[:, ['sha256']]
appsDF['clusterID'] = clusteringLabels
```

```python
# Save the result
appsDF.to_csv(OUTPUT_PATH,index=False)
appsDF.head(3)
```
